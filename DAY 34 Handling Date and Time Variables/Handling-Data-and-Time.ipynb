```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

```python
date['date_month'] = date['date'].dt.month
date.head()
```

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
date = date.drop(columns=['day_pf_week'])
```

```python
date.head()
```

```python
# day of week - name

date['day_of_week_name'] =  date['date'].dt.day_name()
date.head()
```

```python
# is weekend?
date['is_weekend'] = np.where(date['day_of_week_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

```python
#Extract week of the year
date['date_week'] = date['date'].dt.isocalendar().week


date.head()
```

```python
#Extract Quarter
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extract Time elapsed between dates
```

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

date['months_passed'] = (
    (today.year - date['date'].dt.year) * 12 +
    (today.month - date['date'].dt.month)
)

date.head()
```
