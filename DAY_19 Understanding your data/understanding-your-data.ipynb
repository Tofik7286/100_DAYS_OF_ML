Understanding your data 

Once you get the data, what are the first few things you should do?

That's the funda of today's topic.

We have to perform following operation on our data so we can get idea from it.

```python
import pandas as pd 
```

```python
# How big data is?

df  = pd.read_csv("fifa_data.csv")
df.shape
```

```python
# how does data data look like?

df.sample(10)
```

```python
# What are data types of each column?

df.info()
```

```python
# are there any missing values?

df.isnull().sum()
```

```python
# if it is then how much percentage of data is missing?

df.isnull().mean() * 100
```

```python
# how data looks like mathematically?

df.describe()
```

```python
# are there any duplicated rows?
df.duplicated().sum()
```

```python
# how's corelation between different features (features means columns in the dataset)?

df.corr(numeric_only=True)
```
