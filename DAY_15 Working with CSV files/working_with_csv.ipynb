#### 10 Encoding Parameter

```python
# Generally encoding parameter of data is utf-8 but sometimes it is not the case and we have to use other encoding like latin-1, iso-8859-1 etc. like emoji dataset and other language dataset so to handle this we can use encoding parameter


pd.read_csv('zomato.csv').head()
```

```python
# this error we get "UnicodeDecodeError Traceback (most recent call last" because of encoding issue so to handle this we can use encoding parameter

pd.read_csv('zomato.csv',encoding='latin-1').head()
```

#### 11 skip bad lines

```python
# sometime in dataset some rows have more values than the number of columns so to handle this we can use error bad lines parameter

# for example in zomato dataset some rows have more values than the number of columns so to handle this we can use error bad lines parameter

pd.read_csv('zomato.csv', encoding='latin-1', on_bad_lines='warn').head()
```

#### dtypes parameter

```python
# when we use read_csv function it automatically detect data types of columns but sometime it is not the case and we have to specify data types of columns so to handle this we can use dtype parameter 

# for example in our 'aug_train.csv dataset target column has only 0 and 1 values but it is in float data type so to handle this we can use dtype parameter and convert it to int data type

# Syntax : dtype = {'column_name':data_type}

pd.read_csv('aug_train.csv',dtype={'target':int}).info()
```

#### 13 Handling Dates

```python
# when we have date column in our dataset it is in object data type so to handle this we can use parse date parameter and convert it to datetime data type

# for example in our 'IPL Matches 2008-2020.csv' we have date column in object data type so to handle this we can use parse date parameter and convert it to datetime data type

pd.read_csv("IPL Matches 2008-2020.csv").info()

# date             816 non-null    object 
```

```python
# now we use parse date parameter and convert it to datetime data type

# syntax : parse_dates = ['column_name']
pd.read_csv("IPL Matches 2008-2020.csv",parse_dates=['date']).info()

# output would be like this
# 2   date             816 non-null    datetime64[ns]


# Now we can perform all the date time operations on date column like extracting year, month, day, day name, month name etc. because it is in datetime data type now


# if we have like month , day ,and year columns in our dataset and we want to combine them and convert them to datetime data type so to handle this we can use list of lists in parse date parameter

# syntax [[columns_numbers]]

# example : pd.read_csv("dataset_name",parse_dates=[[1,2,3]]).info()
```

#### 14  convertors

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
```

```python
# when we want to apply some transformation on our dataset before loading it to our dataframe we can use converters parameter

# for example in our IPL Matches 2008-2020 dataset we have team1 and team2 columns which have names like this "Royal Challengers Bangalore" and "Sunrisers Hyderabad" but i want like this "RCB" and "SRH" so to handle this we can use converters parameter and apply lambda function to extract first word from team1 and team2 columns

pd.read_csv("IPL Matches 2008-2020.csv",converters={
    'team1' : rename
}).head()
```

#### 15 na_values parameter

```python
# na_values parameter is used to handle missing values in our dataset
# for example in our 'aug_train.csv' dataset we have some missing values in city column which are represented by -1 so to handle this we can use na_values parameter and replace -1 with NaN

pd.read_csv('aug_train.csv',na_values=['city_21'])
```

#### 16 Loading a huge dataset in chunks

```python
# sometime we get a huge dataset and our machine RAM is not sufficient to handle it so to handle this we can use chunksize parameter and read the dataset in chunks.
# we can specify the number of rows we want to read in each chunk using chunksize parameter 
# to access each chunk we can use for loop and iterate through each chunk and perform operations on it


dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
for chunks in dfs:
    print(chunks.shape)
```
