# Iterative Imputation

We perform **Iterative Imputation** using the **MICE Algorithm**.

**MICE** stands for **Multiple Imputation by Chained Equations**.  
It is a powerful technique for handling missing data in datasets.

---

## When to Use?

Before using MICE, we should know about the type of missing data in our dataset.

There are three types of missing data:

### 1. Missing Completely at Random (MCAR)

In this case, the data is missing completely at random, and there is no relationship between the missing data and any other variables in the dataset.  

**Example:** If someone skips a question in a survey by mistake, it is considered MCAR.

### 2. Missing at Random (MAR)

In this case, the data is missing at random, and there is a relationship between the missing data and other variables in the dataset.  

**Example:** If someone skips a question in a survey because they are not comfortable answering it, it is considered MAR.

### 3. Missing Not at Random (MNAR)

In this case, the data is missing not at random, and there is a relationship between the missing data and the variable that is missing.  

**Example:** If someone deletes the data because they don't want to share it, it is considered MNAR.

---

MICE is particularly useful when the missing data is **MAR**.  
It can also be used for **MCAR** and **MNAR**, but it is most effective when the missing data is **MAR**.

---

## Advantages of MICE

- It is accurate.

---

## Disadvantages of MICE

1. It is computationally expensive, especially for large datasets with many missing values.
2. We have to put our entire training dataset on a server to perform MICE, which can be costly.
3. It is slow because it requires multiple iterations to converge to a solution.
