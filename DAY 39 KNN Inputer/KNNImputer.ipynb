# KNN Imputer

- It works on the algorithm of **KNN (K Nearest Neighbors)** to impute missing values in a dataset.

- We fill the missing values with the nearest similar neighbors found in the dataset.

- So how is the similarity calculated?

- The similarity is calculated using distance metrics such as **Euclidean distance**.

---

## It is a 2-step process:

1. Find the **K nearest neighbors** of the data point using Euclidean distance.
2. Impute the missing values by taking the **mean of the K nearest neighbors**.

```python
In KNN k means No. of neighbours. It means how much data points we want to consider to make a descision about the class of a new data point. 
The value of k can be chosen based on the dataset and the problem at hand. 

A common approach is to use cross-validation to find the optimal value of k that gives the best performance on the validation set. 

Generally, a smaller value of k can lead to a more flexible model that may capture noise in the data, while a larger value of k can lead to a smoother decision boundary but may miss important patterns in the data.


For example , if k = 1 then the algorithm will fill the value of the nearest neighbor to the new data point. If k = 3 then algorithm will take mean of the 3 nearest neighbors to fill the value of the new data point.


So now question is how do we caculate the distance between the new data point and the existing data points in the dataset to find the nearest neighbors?

We use this formula to calculate the distance between two data points:

If there is 2D cordinates then we can use the Euclidean distance formula:

distance = sqrt((x2 - x1)^2 + (y2 - y1)^2)

Where (x1, y1) and (x2, y2) are the coordinates of the two data points.

If there are more than 2 dimensions, we can use the generalization of the Euclidean distance formula:

distance = sqrt((x2 - x1)^2 + (y2 - y1)^2 + (z2 - z1)^2 + ...)


Now what if there is  missing values at the neighboring data points? In this case we use nan_euclidean distance which is a extend version of the Euclidean distance that can handle missing values.

The formula for nan_euclidean distance is:
    
    distance = sqrt(weight * (x2 - x1)^2 + (y2 - y1)^2)

weight = number of non-missing values / total number of values

```
