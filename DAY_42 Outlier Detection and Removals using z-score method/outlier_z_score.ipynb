```python
# Outlier removal using z-score : 
# If we want to remove outliers using z-score method then our column must be normally distributed.
# Z-score is calculated as (X - mean) / std deviation
# If the z-score is greater than 3 or less than -3 then we can consider it as an outlier.


# Now how to treat outliers.

# There are 2 popular methods to treat outliers :

# 1. Trimming :
# -> In this method we simply remove the outliers from our dataset. This method is simple and effective but it can lead to loss of data if we have a lot of outliers.

# 2. Capping:
# -> In this method we replace the outliers with a certain value. This value can be the maximum or minimum value of the column or it can be a value that is calculated using some method like IQR (Interquartile Range) method. This method is useful when we have a lot of outliers and we don't want to lose data.

```

```python
print("Highest value allowed",df['cgpa'].mean() + 3 * df['cgpa'].std())
print("Lowest value allowed",df['cgpa'].mean() - 3 * df['cgpa'].std())
```

```python
# finding outliers 

df[(df['cgpa'] > 8.80) | (df['cgpa'] < 5.11)]
```

```python
new_df = df[(df['cgpa'] <= 8.80) & (df['cgpa'] >= 5.11)]
new_df.shape
```

```python
## approach 2 : using z score

df['z_score_CGPA'] = ((df['cgpa']- df['cgpa'].mean()) / df['cgpa'].std())
```

```python
df['z_score_CGPA']
```

```python
df[df['z_score_CGPA'] > 3]
```

```python
df[df['z_score_CGPA'] < -3]
```

```python
df[(df['z_score_CGPA'] > 3 ) | (df['z_score_CGPA'] < -3)]
```

```python
new_df = df[(df['z_score_CGPA'] < 3 ) & (df['z_score_CGPA'] > -3)]
```

```python
new_df.shape
```

#  Capping 

```python
upper_limit = df['cgpa'].mean() + 3 * df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3 * df['cgpa'].std()
```

```python
df['cgpa'] = np.where(
    df['cgpa'] > upper_limit,upper_limit,
    np.where(
        df['cgpa'] < lower_limit,lower_limit,df['cgpa']
    )

)
```
