```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head(5)
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title("Passenger Traveling with")
```

```python
#extract numerical part
df['numeric_number'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
# extract categorical part
df['number_categorical'] = np.where(df['numeric_number'].isnull(),df['number'],np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # capture numeric parts
df['cabin_cat'] = df['Cabin'].str[0] # captures the first letter
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['cabin_cat'].unique()
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
