```python
import pandas as pd 
import numpy as np 
import requests
from bs4 import BeautifulSoup
```

```python
url = f'https://www.ambitionbox.com/it-services-and-consulting-companies-in-india?page=5'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get(url,headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all('h1')[0].text.strip()
```

```python
company = soup.find_all('div',class_='companyCardWrapper')
```

```python
len(company)
```

```python
names=[]
ratings = []
reviews=[]
for i in company:
    names.append(i.find('h2',class_='companyCardWrapper__companyName').text.strip())
    ratings.append(i.find('span',class_='companyCardWrapper__companyRatingCount').text.strip())
    info = i.find_all('div',class_='companyCardWrapper__tertiaryInformation')[0].text.strip()
    reviews.append(info.split('Reviews')[0])
```
