```python
company = soup.find_all('div',class_='companyCardWrapper')[0]
```

```python
company.text
```

```python
name=[]
rating=[]
reviews=[]
salary = []
jobs=[]
benefits = []
photos = []
for i in company:
    name.append(i.find('h2',class_='companyCardWrapper__companyName').text.strip())
    rating.append(i.find('div',class_='rating_text').text.strip())
    info = i.find_all('div',class_='companyCardWrapper__tertiaryInformation')[0].text.strip()
    reviews.append(info.split('Reviews')[0].strip())
    salary.append(info.split('Salaries')[0].split('Reviews')[-1].strip())
    jobs.append(info.split('Jobs')[0].split('Interviews')[-1].strip())
    benefits.append(info.split('Benefits')[0].split('Jobs')[-1].strip())
    photos.append(info.split('Photos')[0].split('Benefits')[-1].strip())
```

```python
df = {
    'name' : name,
    'rating' : rating,
    'reviews':reviews,
    'salary' : salary,
    'jobs' : jobs,
    'benefits' : benefits,
    'photos':photos
}
```

```python
df
```

```python
df = pd.DataFrame(df)
```

```python
final_df = pd.DataFrame()
for j in range(1,10):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    url = f'https://www.ambitionbox.com/it-services-and-consulting-companies-in-india?page={j}'
    webpage = requests.get(url,headers=headers).text
    soup = BeautifulSoup(webpage,'lxml')
    company = soup.find_all('div',class_='companyCardWrapper')
    name=[]
    rating=[]
    reviews=[]
    salary = []
    jobs=[]
    benefits = []
    photos = []
    for i in company:
        try:
            name.append(i.find('h2',class_='companyCardWrapper__companyName').text.strip())
        except:
            name.append(np.nan)
        try:
            rating.append(i.find('div',class_='rating_text').text.strip())
        except:
            rating.append(np.nan)
        info = i.find_all('div',class_='companyCardWrapper__tertiaryInformation')[0].text.strip()
        try:
            reviews.append(info.split('Reviews')[0].strip())
        except:
            reviews.append(np.nan)
        try:
            salary.append(info.split('Salaries')[0].split('Reviews')[-1].strip())
        except:
            salary.append(np.nan)
        try:
            jobs.append(info.split('Jobs')[0].split('Interviews')[-1].strip())
        except:
            jobs.append(np.nan)
        try:
            benefits.append(info.split('Benefits')[0].split('Jobs')[-1].strip())
        except:
            benefits.append(np.nan)
        try:
            photos.append(info.split('Photos')[0].split('Benefits')[-1].strip())
        except:
            photos.append(np.nan)
    d = {
        'name' : name,
        'rating' : rating,
        'reviews':reviews,
        'salary' : salary,
        'jobs' : jobs,
        'benefits' : benefits,
        'photos':photos
    }
    df = pd.DataFrame(d)

    final_df = pd.concat([final_df,df],ignore_index=True)
```
